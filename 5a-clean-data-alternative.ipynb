```python
import pandas as pd
import json
import re
import numpy as np
```

```python
with open('data/store-alternative.json', 'r', encoding='utf8') as f:
    data = json.load(f)
    df = pd.DataFrame.from_dict(data, orient='index')

df.head()
```

```python
df = df.replace(r'^\s*$', None, regex=True)
df.head()
```

```python
df.isna().sum()
```

```python
df = df.fillna('')
```

```python
df.to_csv('data/data-unclean-alternative.csv')
```

```python
dfc = pd.DataFrame()
```

```python
def get_name_bn(row):
    name = row.name.replace(',', '').replace('.txt', '')
    name = re.sub(r'\d+', '', name)
    return name

dfc['name_bn'] = df.apply(get_name_bn, axis=1)
dfc.head()
```

```python
dfc['zillaID'] = df.apply(lambda row: row.name.split(',')[0], axis=1)
dfc['constituencyID'] = df.apply(lambda row: row.name.split(',')[1], axis=1)
dfc.head()
```

```python
def get_tin(row):
    text_matches = re.findall('\d+', row['text_tin'])
    cert_matches = re.findall('\d+', row['cert_tin'])
    for tmatch in text_matches:
        for cmatch in cert_matches:
            if tmatch == cmatch:
                return tmatch
    for match in cert_matches:
        if len(match) == 12:
            return match
    for i, match in enumerate(text_matches):
        if len(match) == 12:
            return match
        if len(match) == 10:
            if i+1 < len(text_matches):
                if text_matches[i+1] and len(text_matches[i+1]) == 2:
                    return match + text_matches[i+1]
            if i+2 < len(text_matches):
                if text_matches[i+1] and len(text_matches[i+1]) == 1:
                    if text_matches[i+2] and len(text_matches[i+2]) == 1:
                        return match + text_matches[i+1] + text_matches[i+2]
    joined = ''.join(text_matches)
    if len(joined) == 12 or len(joined) == 10:
        return joined
    return None

dfc['tin'] = df.apply(get_tin, axis=1)
dfc.head()
```

```python
def get_father(x):
    x = re.sub(r'[^a-z]+', ' ', x)
    for j in junk_words:
        x = x.replace(j, '')
    if x.strip() == '':
        return None
    return x

dfc['father'] = df['cert_father'].apply(get_father)
dfc.head()
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df['cert_mother'].str.split().explode().value_counts())
```

```python
def get_mother(x):
    x = re.sub(r'[^a-z]+', ' ', x)
    for j in junk_words:
        x = x.replace(j, '')
    if x.strip() == '':
        return None
    return x

dfc['mother'] = df['cert_mother'].apply(get_mother)
dfc.head()
```

```python
def get_nid(row):
    text_matches = re.findall('\d+', row['text_nid'])
    slip_matches = re.findall('\d+', row['slip_nid'])
    for match in slip_matches:
        if len(match) == 10:
            return match
    for i, match in enumerate(text_matches):
        if len(match) == 10:
            return match
    joined = ''.join(text_matches)
    if len(joined) == 10:
        return joined
    return None

dfc['nid'] = df.apply(get_nid, axis=1)
dfc.head()
```

```python
def get_current_addr(x):
    for j in junk_words:
        x = x.replace(j, '')
    if x.strip() == '':
        return None
    if len(x) < 10:
        return None
    return x

dfc['current_addr'] = df['cert_current_addr'].apply(get_current_addr)
dfc.head()
```

```python
def get_permanent_addr(x):
    for j in junk_words:
        x = x.replace(j, '')
    if x.strip() == '':
        return None
    if len(x) < 10:
        return None
    return x

dfc['permanent_addr'] = df['cert_permanent_addr'].apply(get_permanent_addr)
dfc.head()
```

```python
count = 0
for i, row in df.iterrows():
    # slip paid
    if row['slip_paid'] and (int(row['slip_paid']) < 100 and int(row['slip_paid']) != 0):
        df.at[i, 'slip_paid'] = ''
        count += 1
    if row['slip_paid'] and int(row['slip_paid']) > 10000000:
        df.at[i, 'slip_paid'] = ''
        count += 1
    if row['slip_paid'] and re.match(r'202[23456]$', row['slip_paid']):
        df.at[i, 'slip_paid'] = ''
        count += 1
    
    # slip shown
    if row['slip_shown'] and int(row['slip_shown']) < 1000:
        df.at[i, 'slip_shown'] = ''
        count += 1
    if row['slip_shown'] and int(row['slip_shown']) > 100000000:
        df.at[i, 'slip_shown'] = ''
        count += 1
    if row['slip_shown'] and re.match(r'202[23456]$', row['slip_shown']):
        df.at[i, 'slip_shown'] = ''
        count += 1
                              
    # cert paid
    if row['cert_paid'] and (int(row['cert_paid']) < 100 and int(row['cert_paid']) != 0):
        df.at[i, 'cert_paid'] = ''
        count += 1
    if row['cert_paid'] and int(row['cert_paid']) > 10000000:
        df.at[i, 'cert_paid'] = ''
        count += 1
    if row['cert_paid'] and re.match(r'202[23456]$', row['cert_paid']):
        df.at[i, 'cert_paid'] = ''
        count += 1
    
    # cert shown
    if row['cert_shown'] and int(row['cert_shown']) < 1000:
        df.at[i, 'cert_shown'] = ''
        count += 1
    if row['cert_shown'] and int(row['cert_shown']) > 100000000:
        df.at[i, 'cert_shown'] = ''
        count += 1
    if row['cert_shown'] and re.match(r'202[23456]$', row['cert_shown']):
        df.at[i, 'cert_shown'] = ''
        count += 1
                              
    # cert wealth
    if row['cert_wealth'] and int(row['cert_wealth']) < 5000:
        df.at[i, 'cert_wealth'] = ''
        count += 1
    if row['cert_wealth'] and int(row['cert_wealth']) > 5000000000:
        df.at[i, 'cert_wealth'] = ''
        count += 1
    if row['cert_wealth'] and re.match(r'202[23456]$', row['cert_wealth']):
        df.at[i, 'cert_wealth'] = ''
        count += 1
        
print(count)
```

```python
count = 0
for i, row in df.iterrows():
    # slip
    if row['slip_paid'] and row['slip_shown']:
        if int(row['slip_paid']) > int(row['slip_shown']):
            df.at[i, 'slip_paid'] = ''
            df.at[i, 'slip_shown'] = ''
            count += 1
    
    # cert
    if row['cert_paid'] and row['cert_shown']:
        if int(row['cert_paid']) > int(row['cert_shown']):
            df.at[i, 'cert_paid'] = ''
            df.at[i, 'cert_shown'] = ''
            df.at[i, 'cert_wealth'] = ''
            count += 1
    if row['cert_paid'] and row['cert_wealth']:
        if int(row['cert_paid']) > int(row['cert_wealth']):
            df.at[i, 'cert_paid'] = ''
            df.at[i, 'cert_shown'] = ''
            df.at[i, 'cert_wealth'] = ''
            count += 1
    if row['cert_wealth'] and row['cert_shown']:
        if int(row['cert_wealth']) < int(row['cert_shown']):
            df.at[i, 'cert_paid'] = ''
            df.at[i, 'cert_shown'] = ''
            df.at[i, 'cert_wealth'] = ''
            count += 1

print(count)
```

```python
def get_shown(row):
    text_matches = re.findall('\d{4,}', row['text_shown'])
    slip_matches = re.findall('\d{4,}', row['slip_shown'])
    cert_matches = re.findall('\d{4,}', row['cert_shown'])
    for cmatch in cert_matches:
        if cmatch in text_matches:
            return cmatch
    for smatch in slip_matches:
        if smatch in text_matches:
            return smatch
    if 'net wealth' in row['text_shown']:
        idx = row['text_shown'].index('net wealth')
        for match in text_matches:
            if row['text_shown'].index(match) < idx:
                return match
    if 'tax paid' in row['text_shown']:
        idx = row['text_shown'].index('tax paid')
        for match in text_matches:
            if row['text_shown'].index(match) < idx:
                return match
    return None

dfc['shown'] = df.apply(get_shown, axis=1)
dfc.head()
```

```python
def get_paid(row):
    text_matches = re.findall('\d+', row['text_paid'])
    slip_matches = re.findall('\d+', row['slip_paid'])
    cert_matches = re.findall('\d+', row['cert_paid'])
    for cmatch in cert_matches:
        if cmatch in text_matches:
            return cmatch
    for smatch in slip_matches:
        if smatch in text_matches:
            return smatch
    wtext_matches = re.findall(r'paid tax \d+', row['text_wealth'])
    if len(wtext_matches) > 0:
        return wtext_matches[0].split(' ')[-1].strip()
    if len(text_matches) > 0:
        return text_matches[-1]
    return None

dfc['paid'] = df.apply(get_paid, axis=1)
dfc.head()
```

```python
def get_wealth(row):
    text_matches = re.findall('\d+', row['text_wealth'])
    cert_matches = re.findall('\d+', row['cert_wealth'])
    for cmatch in cert_matches:
        if cmatch in text_matches:
            if int(cmatch) > 10000:
                return cmatch
    pattern_matches = re.findall(r'\d+ [a-z]+ \d+ [a-z]+ \d+', row['text_wealth'])
    if len(pattern_matches) > 0:
        matches = re.findall('\d+', pattern_matches[0])
        if int(matches[1]) == max([int(m) for m in matches]):
            if int(matches[1]) > 10000:
                return matches[1]
    return None

dfc['wealth'] = df.apply(get_wealth, axis=1)
dfc.head()
```
