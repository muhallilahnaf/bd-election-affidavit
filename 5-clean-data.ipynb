```python
import pandas as pd
import json
import re
```

```python
with open('data/store.json', 'r', encoding='utf8') as f:
    data = json.load(f)
    df = pd.DataFrame.from_dict(data, orient='index')

df.head()
```

```python
df.isna().sum()
```

```python
df.to_csv('data/data_unclean.csv')
```

```python
dfc = pd.DataFrame()
```

```python
def get_name_bn(row):
    name = row.name.replace(',', '').replace('.txt', '')
    name = re.sub(r'\d+', '', name)
    return name

dfc['name_bn'] = df.apply(get_name_bn, axis=1)
dfc.head()
```

```python
dfc['zillaID'] = df.apply(lambda row: row.name.split(',')[0], axis=1)
dfc['constituencyID'] = df.apply(lambda row: row.name.split(',')[1], axis=1)
dfc.head()
```

```python
def get_text_tin(x):
    matches = re.findall('\d+', x)
    for i, match in enumerate(matches):
        if len(match) == 12:
            return match
        if len(match) == 10:
            if i+1 < len(matches):
                if matches[i+1] and len(matches[i+1]) == 2:
                    return match + matches[i+1]
            if i+2 < len(matches):
                if matches[i+1] and len(matches[i+1]) == 1:
                    if matches[i+2] and len(matches[i+2]) == 1:
                        return match + matches[i+1] + matches[i+2]
    joined = ''.join(matches)
    if len(joined) == 12 or len(joined) == 10:
        return joined
    return None

dfc['text_tin'] = df['text_tin'].apply(get_text_tin)
dfc.head()
```

```python
def get_text_father(x):
    if ' tin ' in x:
        end = x.index(' tin ')
        x = x[:end]
    x = re.sub(r'[^a-z]+', ' ', x)
    if x.strip() == '':
        return None
    return x

dfc['text_father'] = df['text_father'].apply(get_text_father)
dfc.head()
```

```python
def get_text_mother(x):
    if ' tin ' in x:
        return None
    x = re.sub(r'[^a-z]+', ' ', x)
    if x.strip() == '':
        return None
    return x

dfc['text_mother'] = df['text_mother'].apply(get_text_mother)
dfc.head()
```

```python
def get_text_current_addr(x):
    x = x.lower()
    if 'permanent' in x:
        x = x.split('permanent')[0]
    x = x.replace('current', '').replace('permanent', '').replace('address', '')
    x = re.sub(r'4.{0,2}b.{0,2}\)', '', x)
    x = re.sub(r'4.{0,2}a.{0,2}\)', '', x)
    x = re.sub(r'^\W+', '', x)
    return x.strip()

dfc['text_current_addr'] = df['text_current_addr'].apply(get_text_current_addr)
dfc.head()
```

```python
def get_text_permanent_addr(x):
    x = x.lower()
    x = x.replace('permanent', '').replace('address', '')
    x = re.sub(r'4.{0,2}b.{0,2}\)', '', x)
    x = re.sub(r'4.{0,2}a.{0,2}\)', '', x)
    x = re.sub(r'^\W+', '', x)
    return x.strip()

dfc['text_permanent_addr'] = df['text_permanent_addr'].apply(get_text_permanent_addr)
dfc.head()
```

```python
def get_text_shown(x):
    x = re.sub(r'[\.\,\s]+', '', x)
    if 'shown' not in x:
        return None
    shown_idx = x.index('shown')
    target_idx = None
    target_match = ''
    matches = re.findall('\d+', x)
    for match in matches:
        idx = x.index(match)
        if len(match) > 3 and shown_idx < idx:
            if not target_idx:
                target_idx = idx
                target_match = match
            else:
                if (idx-shown_idx) < (target_idx-shown_idx):
                    target_idx = idx
                    target_match = match
    return target_match

dfc['text_shown'] = df['text_shown'].apply(get_text_shown)
dfc.head()
```

```python
def get_text_wealth(x):
    x = re.sub(r'[\.\,\s]+', '', x)
    if 'wealth' not in x:
        return None
    wealth_idx = x.index('wealth')
    target_idx = None
    target_match = ''
    matches = re.findall('\d+', x)
    for match in matches:
        idx = x.index(match)
        if len(match) > 3 and wealth_idx < idx:
            if not target_idx:
                target_idx = idx
                target_match = match
            else:
                if (idx-wealth_idx) < (target_idx-wealth_idx):
                    target_idx = idx
                    target_match = match
    return target_match

dfc['text_wealth'] = df['text_shown'].apply(get_text_wealth)
dfc.head()
```

```python
def get_text_paid(x):
    x = re.sub(r'[\.\,\s]+', '', x)
    if 'paid' not in x:
        return None
    paid_idx = x.index('paid')
    target_idx = None
    target_match = None
    matches = re.findall('\d+', x)
    for match in matches:
        idx = x.index(match)
        if len(match) > 3 and paid_idx < idx:
            if not target_idx:
                target_idx = idx
                target_match = match
            else:
                if (idx-paid_idx) < (target_idx-paid_idx):
                    target_idx = idx
                    target_match = match
    return target_match

dfc['text_paid'] = df['text_paid'].apply(get_text_paid)
dfc.head()
```
